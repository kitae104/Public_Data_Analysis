#### 데이터 분석


```python
import csv

a = open('./SNS이용률.csv', 'r', encoding='utf-8')
data = csv.reader(a)
header = next(data)

del header[:2]

num = 0

sns_facebook = []
for row in data:
    row[2:] = map(float, row[2:]) # 문자열을 실수로 변환
    if num == 5:
        print(row)
        for i in range(2, len(row)):
            sns_facebook.append(row[i])
    num += 1

for i in range(len(sns_facebook)):
    print(header[i], sns_facebook[i])

a.close()
```

```python
import csv
import matplotlib.pyplot as plt

a = open('./SNS이용률.csv', 'r', encoding='utf-8')
data = csv.reader(a)
header = next(data)

del header[:2]

sns_name = []
sns_use = []

sns = ['커뮤니티(카페/클럽등)', '블로그(네이버 다음 등)', '미니홈피', '페이스북', '인스타그램', '트위터', '틱톡(TikTok)']

for row in data:
    if row[1].strip() in sns:
        row[2:] = map(float, row[2:]) # 문자열을 실수로 변환
        sns_name.append(row[1])
        sns_use.append(row[2:])

plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 5))
for i in range(len(sns_name)):
    plt.plot(header, sns_use[i], label=sns_name[i])
plt.title('SNS 이용률')
plt.xlabel('연도')
plt.ylabel('이용률(%)')
plt.grid()
plt.legend( loc='upper left', fontsize='small' )
plt.show()

a.close()
```
