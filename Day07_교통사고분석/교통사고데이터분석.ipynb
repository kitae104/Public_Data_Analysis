```python
from calendar import weekday
import csv
import matplotlib.pyplot as plt

a = open("./data.csv", "r")
data = csv.reader(a)

num = 0
week = [0] * 7
weekday = ["일", "월", "화", "수", "목", "금", "토"]
print(week)

for i in range(5):
    next(a)  # 헤더 제거

for row in data:
    row[3:] = map(int, row[3:])
    if num % 3 == 0:
        for i in range(7):
            week[i] += row[3 + i]
    num += 1

for i in range(len(week)):
    print(f"{weekday[i]} : {week[i]}")

plt.rc("font", family="Malgun Gothic")
plt.plot(weekday, week, "r--")
plt.show()

a.close()
```

```python
import csv

a = open("./data.csv", "r")
data = csv.reader(a)

num = 0
max = 0

for i in range(5):
    next(a)  # 헤더 제거

for row in data:
    row[3:] = map(int, row[3:])
    if num % 3 == 0:
        if max < row[8]:
            max = row[8]
            name = row[0]
    num += 1

print(f"금요일에 사고가 가장 많이 발생한 지역은 {name}입니다.")

a.close()
```

#### 그래프 표시 및 분석

```python
import csv

a = open("./data.csv", "r")
data = csv.reader(a)

num = 0
ac = []
name = []

for i in range(5):
    next(a)  # 헤더 제거

for row in data:
    if num % 3 == 0:
        name.append(row[0])
        ac.append(int(row[8]))

    num += 1

plt.rc("font", family="Malgun Gothic")
plt.axis("equal")
plt.pie(ac, labels=name, autopct="%1.1f%%")
plt.show()

a.close()
```
