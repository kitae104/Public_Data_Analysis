- 행안부 : www.mois.go.kr
- 정책자료 → 통계 → 주민등록 인구통계
- 연령별 인구현황 - 1세 단위 - 0~ 100세 (age, gender) 파일 생성 

```python
import csv


f = open("./test1.csv")
data = csv.reader(f)

result = []                     # 인구 정보를 저장할 리스트

findDong = input("동 이름 : ")

for row in data:
    # print(row)                # 전체 확인
    if findDong in row[0]:
        #print(row)             # 해당 동 정보 확인
        for i in row[3:]:       # 0세 ~ 100세
            #print(i)
            # 리스트에 인구수 추가(,가 있는 경우 제거한후 숫자로 변경)
            result.append(int(i.replace(',','')))
print(result)

# 시각화
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')     # 한글 폰트 적용
plt.style.use('ggplot')                         # 격자 무늬 스타일 지정
plt.plot(result)
plt.title(findDong + " 인구 구조")
plt.show()
```
