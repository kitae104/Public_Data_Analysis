```python
# 남여 인구 구조 시각화 - 항아리 모양 그래프
import csv
f = open("./test2.csv")
data = csv.reader(f)

m = []              # 남성 정보 리스트
f = []              # 여성 정보 리스트

findDong = input("동 이름 : ")

for row in data:
    if findDong in row[0]:
        for i in row[3:104]:                    # 남성 데이터
            m.append(-int(i))
        for i in row[106:]:                     # 여성 데이터
            f.append(int(i))

print(m)
print(f)


import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')        # 한글 폰트 적용
plt.rcParams['axes.unicode_minus'] = False      # 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
plt.title(findDong + " 지역의 남녀 성별 인구 분포")
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()
```
