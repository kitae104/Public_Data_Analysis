#### 데이터 분석


```python
import csv

a = open('./범죄발생_시간.csv', 'r')
data = csv.reader(a) 
header = next(data) #  시간대 정보 

max = 0
count = 0
i = 0

crime_name = [0] * 31
crime_maxcount = [0] * 31
crime_time = [0] * 31

for row in data:
    row[1:] = map(int, row[1:])
    max = 0
    count = 0
    
    for num in range(1, 10):
        if max < row[num]:
            max = row[num]
            count = num
            
    crime_time[i] = header[count]
    crime_name[i] = row[0]
    crime_maxcount[i] = max
    i += 1
    
for i in range(len(crime_name)):
    print(f"{crime_time[i]}에 {crime_name[i]}범죄가 최대 {crime_maxcount[i]}건 발생했으니 조심하세요.")
    
a.close()
```

#### 그래프 표시 및 분석


```python
import csv
from tkinter import font
import matplotlib.pyplot as plt

a = open('./범죄발생_시간.csv', 'r')
data = csv.reader(a) 
header = next(data) #  시간대 정보 
del(header[0]) # 범죄명 제거

count = 0

crime_name = [0] * 31
crime_count = [0] * 31

for row in data:
    row[1:] = map(int, row[1:])
    
    crime_name[count] = row[0]
    crime_count[count] = row[1:]
    count += 1
    
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15, 7))
plt.title('시간대별 범죄 발생 건수')
for i in range(len(crime_name)):
    plt.plot(header, crime_count[i], label=crime_name[i])
    
plt.xticks(rotation=45)    
plt.legend(fontsize=7, loc='upper left', ncol=4)
plt.show()
a.close()
```
