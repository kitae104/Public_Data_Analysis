#### 매수/매도 호가 정보

```python
import pybithumb

order = pybithumb.get_orderbook("BTT")
bids = order["bids"]
asks = order["asks"]

for bid in bids:
    price = bid["price"]
    quantity = bid["quantity"]
    print("매수 호가 : ", price, ", 매수 잔량 : ", quantity)

print("-" * 50)

for ask in asks:
    price = ask["price"]
    quantity = ask["quantity"]
    print("매도 호가 : ", price, ", 매도 잔량 : ", quantity)
```

#### 5일 평균 구하기

```python
import pybithumb
ohlcv = pybithumb.get_ohlcv("BTT")
close = ohlcv['close']
mean5 = close.rolling(window=5).mean()
print(mean5)
```
