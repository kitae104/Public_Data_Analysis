```python
import pybithumb
```

```python
def bull_market(ticker):
    ohlcv = pybithumb.get_ohlcv(ticker)  # 거래소 과거 시세 가져오기
    # 데이터 없음 또는 충분치 않으면 False 반환(방어적 처리)
    if ohlcv is None or len(ohlcv) < 5:
        return False
    close = ohlcv["close"]  # 종가(close) 가져오기
    mean5 = close.rolling(window=5).mean()  # 5일 평균 구하기
    price = pybithumb.get_current_price(ticker)  # 현재가 가져오기
    if price is None:
        return False

    # 위치 기반 인덱싱으로 마지막 전(최근 이전) 평균값을 가져옴
    last_mean5 = mean5.iloc[-2]

    if price > last_mean5:
        return True
    else:
        return False
```

```python
# tickers = pybithumb.get_tickers() # 거래소에서 거래되는 모든 암호화폐 목록 가져오기
tickers = ["BTT", "AVAX", "XRP"]

for ticker in tickers:
    if bull_market(ticker):
        print(ticker, "상승장")
    else:
        print(ticker, "하락장")
```
