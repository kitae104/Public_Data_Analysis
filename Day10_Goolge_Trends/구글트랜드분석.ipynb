```python
from pytrends.request import TrendReq
import plotly.express as px
```

```python
import time
import plotly.io as pio

pytrends = TrendReq(hl='ko', tz=540) # 한국어, 한국 시간대
keyword_list = ['Java', 'Python', 'R', 'LLM', '인공지능'] # 관심 키워드 리스트
# 5년간의 관심도 데이터 요청
pytrends.build_payload(keyword_list, cat=0, timeframe='today 5-y', geo='KR', gprop='youtube')

data = pytrends.interest_over_time() # 관심도 데이터
data = data.drop(columns=['isPartial']) # isPartial 열 제거
data = data.reset_index() # 인덱스 초기화
print(data.head()) # 데이터 확인

figure = px.line(data, x='date', y=keyword_list, title='Java, Python, R, LLM, 인공지능의 관심도 추이')
figure.show(renderer="browser") # 브라우저에서 그래프 표시
```

```python
from pytrends.request import TrendReq
import plotly.io as pio

pytrends = TrendReq(hl='en-US', tz=360) # 영어, 미국 시간대
keyword_list = ['Java', 'Python', 'LLM', 'Artificial Intelligence'] # 관심 키워드 리스트
# 5년간의 관심도 데이터 요청
pytrends.build_payload(keyword_list, cat=0, timeframe='today 5-y', geo='US', gprop='')
data = pytrends.interest_over_time() # 관심도 데이터
data = data.drop(columns=['isPartial']) # isPartial 열 제거
data = data.reset_index() # 인덱스 초기화
print(data.head()) # 데이터 확인

figure = px.line(data, x='date', y=keyword_list, title='Java, Python, LLM, Artificial Intelligence Interest Over Time')
figure.show(renderer="browser") # 브라우저에서 그래프 표시
```

```python
import matplotlib.pyplot as plt
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)

kw_list = ['Java', 'Python', 'LLM', 'Artificial Intelligence'] # 관심 키워드 리스트
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='youtube')
df=pytrends.interest_over_time()
df=df.drop(['isPartial'],axis=1)

plt.figure(figsize=(20,10))
plt.rc('font', size=15) # 폰트 크기 설정
plt.rc("font", family="Malgun Gothic") # 한글 폰트 설정

plt.plot(df.index, df['Java'], label='Java')
plt.plot(df.index, df['Python'], label='Python')
plt.plot(df.index, df['LLM'], label='LLM')
plt.plot(df.index, df['Artificial Intelligence'], label='Artificial Intelligence')
plt.title('Java, Python, LLM, Artificial Intelligence Interest Over Time')
plt.legend()
plt.show()
```
