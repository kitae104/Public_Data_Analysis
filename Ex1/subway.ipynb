```python
import pandas as pd

# Load the data from the uploaded CSV file
file_path = '/mnt/data/인천교통공사_1호선 일별 승하차현황_20240630.csv'
data = pd.read_csv(file_path, encoding='utf-8')

# Display the first few rows of the dataframe to understand its structure
data.head()
```

# Convert the '일자' column to datetime format for easier analysis
data['일자'] = pd.to_datetime(data['일자'])

# Group by date and calculate the total 승차인원 and 하차인원
daily_stats = data.groupby('일자')[['승차인원', '하차인원']].sum().reset_index()

# Sort by 승차인원 and 하차인원 to find top 5 days
top_5_days_by_승차 = daily_stats.nlargest(5, '승차인원')
top_5_days_by_하차 = daily_stats.nlargest(5, '하차인원')

import matplotlib.pyplot as plt

# Plot the top 5 days by 승차인원
plt.figure(figsize=(12, 6))
plt.plot(daily_stats['일자'], daily_stats['승차인원'], label='승차인원')
plt.plot(daily_stats['일자'], daily_stats['하차인원'], label='하차인원')
plt.title('날짜별 승차 및 하차 인원 추이')
plt.xlabel('날짜')
plt.ylabel('인원 수')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

top_5_days_by_승차, top_5_days_by_하차


```python
# Add a new column for month extracted from '일자'
data['월'] = data['일자'].dt.to_period('M')

# Group by month and calculate the total 승차인원 and 하차인원
monthly_stats = data.groupby('월')[['승차인원', '하차인원']].sum().reset_index()

# Sort by 승차인원 and 하차인원 to find top 5 months
top_5_months_by_승차 = monthly_stats.nlargest(5, '승차인원')
top_5_months_by_하차 = monthly_stats.nlargest(5, '하차인원')

# Plot the monthly 승차인원 and 하차인원
plt.figure(figsize=(12, 6))
plt.plot(monthly_stats['월'].astype(str), monthly_stats['승차인원'], marker='o', label='승차인원')
plt.plot(monthly_stats['월'].astype(str), monthly_stats['하차인원'], marker='o', label='하차인원')
plt.title('월별 승차 및 하차 인원 추이')
plt.xlabel('월')
plt.ylabel('인원 수')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

top_5_months_by_승차, top_5_months_by_하차
```

```python
# Group by station and calculate the total 승차인원 and 하차인원
station_stats = data.groupby('역명')[['승차인원', '하차인원']].sum().reset_index()

# Sort by 승차인원 and 하차인원 to find top 5 stations
top_5_stations_by_승차 = station_stats.nlargest(5, '승차인원')
top_5_stations_by_하차 = station_stats.nlargest(5, '하차인원')

# Plot the station 승차인원 and 하차인원 for top 5 stations
plt.figure(figsize=(12, 6))
plt.bar(top_5_stations_by_승차['역명'], top_5_stations_by_승차['승차인원'], label='승차인원', alpha=0.7)
plt.bar(top_5_stations_by_하차['역명'], top_5_stations_by_하차['하차인원'], label='하차인원', alpha=0.7)
plt.title('역별 승차 및 하차 인원 (상위 5개 역)')
plt.xlabel('역명')
plt.ylabel('인원 수')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

top_5_stations_by_승차, top_5_stations_by_하차
```

```python
# Group by station and month, and calculate the total 승차인원 and 하차인원
station_monthly_stats = data.groupby(['역명', '월'])[['승차인원', '하차인원']].sum().reset_index()

# Find the top 5 station-month combinations by 승차인원 and 하차인원
top_5_station_months_by_승차 = station_monthly_stats.nlargest(5, '승차인원')
top_5_station_months_by_하차 = station_monthly_stats.nlargest(5, '하차인원')

# Plot the top 5 station-month combinations by 승차인원 and 하차인원
plt.figure(figsize=(12, 6))
plt.bar(top_5_station_months_by_승차['역명'] + ' ' + top_5_station_months_by_승차['월'].astype(str), 
        top_5_station_months_by_승차['승차인원'], label='승차인원', alpha=0.7)
plt.bar(top_5_station_months_by_하차['역명'] + ' ' + top_5_station_months_by_하차['월'].astype(str), 
        top_5_station_months_by_하차['하차인원'], label='하차인원', alpha=0.7)
plt.title('역별 월별 승차 및 하차 인원 (상위 5개)')
plt.xlabel('역명 월')
plt.ylabel('인원 수')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

top_5_station_months_by_승차, top_5_station_months_by_하차
```

```python
# Find the station with the most and least usage
station_stats['총이용인원'] = station_stats['승차인원'] + station_stats['하차인원']
most_used_station = station_stats.nlargest(5, '총이용인원')
least_used_station = station_stats.nsmallest(5, '총이용인원')

# Plot the most and least used stations
plt.figure(figsize=(12, 6))
plt.bar(most_used_station['역명'], most_used_station['총이용인원'], label='총이용인원 (가장 많은 역)', alpha=0.7)
plt.bar(least_used_station['역명'], least_used_station['총이용인원'], label='총이용인원 (가장 적은 역)', alpha=0.7)
plt.title('이용 인원이 가장 많은 역과 적은 역')
plt.xlabel('역명')
plt.ylabel('총 이용 인원 수')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

most_used_station, least_used_station
```

```python
# Calculate monthly usage per station
station_monthly_stats['총이용인원'] = station_monthly_stats['승차인원'] + station_monthly_stats['하차인원']

# Find the top 5 station-month combinations by total usage
most_used_station_month = station_monthly_stats.nlargest(5, '총이용인원')
least_used_station_month = station_monthly_stats.nsmallest(5, '총이용인원')

# Plot the most and least used station-month combinations
plt.figure(figsize=(12, 6))
plt.bar(most_used_station_month['역명'] + ' ' + most_used_station_month['월'].astype(str), 
        most_used_station_month['총이용인원'], label='총이용인원 (가장 많은 조합)', alpha=0.7)
plt.bar(least_used_station_month['역명'] + ' ' + least_used_station_month['월'].astype(str), 
        least_used_station_month['총이용인원'], label='총이용인원 (가장 적은 조합)', alpha=0.7)
plt.title('월별 이용 인원이 가장 많은 역과 적은 역')
plt.xlabel('역명 월')
plt.ylabel('총 이용 인원 수')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

most_used_station_month, least_used_station_month
```
